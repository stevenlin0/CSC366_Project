```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

df = pd.read_csv('all_mlb_data.csv')

print("=" * 60)
print("STEP 1: DATA CLEANING AND PREPROCESSING")
print("=" * 60)
print(f"\nInitial data shape: {df.shape}")
print(f"\nInitial data info:")
print(df.info())
print(f"\nFirst few rows:")
print(df.head(10))
```

```python
print("\n" + "=" * 60)
print("Cleaning Data: Removing empty rows and separating player types")
print("=" * 60)

df = df.dropna(how='all')

pitcher_start_idx = df[df['Name'].isna() & df['AAV (M)'].isna()].index
if len(pitcher_start_idx) > 0:
    hitters_df = df.iloc[:pitcher_start_idx[0]].copy()
    pitchers_df = df.iloc[pitcher_start_idx[0]+1:].copy()
    
    if pitchers_df.iloc[0]['Name'] == 'Name':
        pitchers_df = pitchers_df.iloc[1:].copy()
else:
    if 'ERA' in df.columns:
        hitters_df = df[df['ERA'].isna()].copy()
        pitchers_df = df[df['ERA'].notna()].copy()
    else:
        hitters_df = df.copy()
        pitchers_df = pd.DataFrame()

hitters_df = hitters_df[hitters_df['Name'].notna()].copy()
if not pitchers_df.empty:
    pitchers_df = pitchers_df[pitchers_df['Name'].notna()].copy()

print(f"\nHitters data shape: {hitters_df.shape}")
print(f"Pitchers data shape: {pitchers_df.shape if not pitchers_df.empty else (0, 0)}")
```

```python
print("\n" + "=" * 60)
print("Handling Missing Values and Data Types - Hitters")
print("=" * 60)

numeric_cols_hitters = ['AAV (M)', 'WAR', 'AVG', 'OBP', 'SLG', 'OPS', 'wRC+']
for col in numeric_cols_hitters:
    if col in hitters_df.columns:
        hitters_df[col] = pd.to_numeric(hitters_df[col], errors='coerce')

print("\nMissing values in hitters data:")
print(hitters_df[numeric_cols_hitters].isnull().sum())

print("\n" + "-" * 60)
print("Outlier Detection - Hitters (using IQR method)")
print("-" * 60)

for col in ['WAR', 'AAV (M)', 'wRC+']:
    if col in hitters_df.columns:
        Q1 = hitters_df[col].quantile(0.25)
        Q3 = hitters_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = hitters_df[(hitters_df[col] < lower_bound) | (hitters_df[col] > upper_bound)]
        print(f"\n{col}:")
        print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
        print(f"  Outliers: {len(outliers)} ({len(outliers)/len(hitters_df)*100:.1f}%)")
        if len(outliers) > 0:
            print(f"  Outlier values: {outliers[['Name', col]].values.tolist()}")

print("\nHitters data summary:")
print(hitters_df.describe())
```

```python
if not pitchers_df.empty:
    print("\n" + "=" * 60)
    print("Handling Missing Values and Data Types - Pitchers")
    print("=" * 60)
    
    numeric_cols_pitchers = ['AAV (M)', 'WAR', 'ERA', 'FIP', 'K/9', 'BB/9', 'BABIP']
    for col in numeric_cols_pitchers:
        if col in pitchers_df.columns:
            pitchers_df[col] = pd.to_numeric(pitchers_df[col], errors='coerce')
    
    print("\nMissing values in pitchers data:")
    print(pitchers_df[numeric_cols_pitchers].isnull().sum())
    
    print("\n" + "-" * 60)
    print("Outlier Detection - Pitchers (using IQR method)")
    print("-" * 60)
    
    for col in ['WAR', 'AAV (M)', 'ERA']:
        if col in pitchers_df.columns:
            Q1 = pitchers_df[col].quantile(0.25)
            Q3 = pitchers_df[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = pitchers_df[(pitchers_df[col] < lower_bound) | (pitchers_df[col] > upper_bound)]
            print(f"\n{col}:")
            print(f"  Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
            print(f"  Outliers: {len(outliers)} ({len(outliers)/len(pitchers_df)*100:.1f}%)")
            if len(outliers) > 0:
                print(f"  Outlier values: {outliers[['Name', col]].values.tolist()}")
    
    print("\nPitchers data summary:")
    print(pitchers_df.describe())
else:
    print("\nNo pitchers data found.")
```

```python
print("\n" + "=" * 60)
print("STEP 2: EXPLORATORY DATA ANALYSIS")
print("=" * 60)

hitters_df['WAR_per_M'] = hitters_df['WAR'] / hitters_df['AAV (M)']
hitters_df['Contract_Value'] = hitters_df['WAR_per_M'].apply(
    lambda x: 'Good' if x >= 0.5 else 'Poor' if x < 0.2 else 'Average'
)

if not pitchers_df.empty:
    pitchers_df['WAR_per_M'] = pitchers_df['WAR'] / pitchers_df['AAV (M)']
    pitchers_df['Contract_Value'] = pitchers_df['WAR_per_M'].apply(
        lambda x: 'Good' if x >= 0.3 else 'Poor' if x < 0.1 else 'Average'
    )

print("\nHitters - Basic Statistics:")
print(f"Total players: {len(hitters_df)}")
print(f"Average AAV: ${hitters_df['AAV (M)'].mean():.2f}M")
print(f"Average WAR: {hitters_df['WAR'].mean():.2f}")
print(f"Average WAR per $M: {hitters_df['WAR_per_M'].mean():.3f}")
print(f"\nContract Value Distribution:")
print(hitters_df['Contract_Value'].value_counts())

if not pitchers_df.empty:
    print("\n\nPitchers - Basic Statistics:")
    print(f"Total players: {len(pitchers_df)}")
    print(f"Average AAV: ${pitchers_df['AAV (M)'].mean():.2f}M")
    print(f"Average WAR: {pitchers_df['WAR'].mean():.2f}")
    print(f"Average WAR per $M: {pitchers_df['WAR_per_M'].mean():.3f}")
    print(f"\nContract Value Distribution:")
    print(pitchers_df['Contract_Value'].value_counts())
```

```python
print("\n" + "-" * 60)
print("Correlation Analysis - Hitters")
print("-" * 60)

hitter_corr_cols = ['AAV (M)', 'WAR', 'AVG', 'OBP', 'SLG', 'OPS', 'wRC+', 'WAR_per_M']
hitter_corr = hitters_df[hitter_corr_cols].corr()

print("\nCorrelation Matrix (Hitters):")
print(hitter_corr.round(3))

print("\nKey Correlations:")
print(f"AAV vs WAR: {hitters_df['AAV (M)'].corr(hitters_df['WAR']):.3f}")
print(f"AAV vs wRC+: {hitters_df['AAV (M)'].corr(hitters_df['wRC+']):.3f}")
print(f"WAR vs OPS: {hitters_df['WAR'].corr(hitters_df['OPS']):.3f}")

if not pitchers_df.empty:
    print("\n" + "-" * 60)
    print("Correlation Analysis - Pitchers")
    print("-" * 60)
    
    pitcher_corr_cols = ['AAV (M)', 'WAR', 'ERA', 'FIP', 'K/9', 'BB/9', 'WAR_per_M']
    pitcher_corr = pitchers_df[pitcher_corr_cols].corr()
    
    print("\nCorrelation Matrix (Pitchers):")
    print(pitcher_corr.round(3))
    
    print("\nKey Correlations:")
    print(f"AAV vs WAR: {pitchers_df['AAV (M)'].corr(pitchers_df['WAR']):.3f}")
    print(f"AAV vs ERA: {pitchers_df['AAV (M)'].corr(pitchers_df['ERA']):.3f} (negative is better)")
    print(f"WAR vs ERA: {pitchers_df['WAR'].corr(pitchers_df['ERA']):.3f} (negative is expected)")
```

```python
print("\n" + "-" * 60)
print("Analysis by Contract Era")
print("-" * 60)

hitters_df['Contract_Start'] = hitters_df['Contract Years'].str.split('-').str[0].astype(int)

print("\nHitters - Performance by Contract Era:")
era_analysis_hitters = hitters_df.groupby('Contract_Start').agg({
    'AAV (M)': 'mean',
    'WAR': 'mean',
    'WAR_per_M': 'mean',
    'wRC+': 'mean'
}).round(3)
print(era_analysis_hitters)

if not pitchers_df.empty:
    pitchers_df['Contract_Start'] = pitchers_df['Contract Years'].str.split('-').str[0].astype(int)
    
    print("\n\nPitchers - Performance by Contract Era:")
    era_analysis_pitchers = pitchers_df.groupby('Contract_Start').agg({
        'AAV (M)': 'mean',
        'WAR': 'mean',
        'WAR_per_M': 'mean',
        'ERA': 'mean'
    }).round(3)
    print(era_analysis_pitchers)
```

```python
print("\n" + "=" * 60)
print("STEP 3: STATISTICAL ANALYSIS")
print("=" * 60)

print("\n" + "-" * 60)
print("Statistical Test 1: Correlation between AAV and WAR")
print("-" * 60)

hitter_aav_war_corr, hitter_p_value = stats.pearsonr(
    hitters_df['AAV (M)'].dropna(), 
    hitters_df['WAR'].dropna()
)
print(f"\nHitters:")
print(f"  Pearson correlation coefficient: {hitter_aav_war_corr:.3f}")
print(f"  P-value: {hitter_p_value:.4f}")
if hitter_p_value < 0.05:
    print(f"  Result: Significant relationship (p < 0.05)")
else:
    print(f"  Result: No significant relationship (p >= 0.05)")

if not pitchers_df.empty:
    pitcher_aav_war_corr, pitcher_p_value = stats.pearsonr(
        pitchers_df['AAV (M)'].dropna(), 
        pitchers_df['WAR'].dropna()
    )
    print(f"\nPitchers:")
    print(f"  Pearson correlation coefficient: {pitcher_aav_war_corr:.3f}")
    print(f"  P-value: {pitcher_p_value:.4f}")
    if pitcher_p_value < 0.05:
        print(f"  Result: Significant relationship (p < 0.05)")
    else:
        print(f"  Result: No significant relationship (p >= 0.05)")
```

```python
print("\n" + "-" * 60)
print("Statistical Test 2: Performance Differences Across Eras")
print("-" * 60)

hitters_df['Era'] = hitters_df['Contract_Start'].apply(
    lambda x: 'Recent (2022+)' if x >= 2022 else 'Mid (2015-2018)' if x >= 2015 else 'Early (2008-2012)'
)

print("\nHitters - WAR by Era:")
era_groups_hitters = hitters_df.groupby('Era')['WAR']
print(era_groups_hitters.describe())

recent_hitters = hitters_df[hitters_df['Era'] == 'Recent (2022+)']['WAR'].dropna()
mid_hitters = hitters_df[hitters_df['Era'] == 'Mid (2015-2018)']['WAR'].dropna()
early_hitters = hitters_df[hitters_df['Era'] == 'Early (2008-2012)']['WAR'].dropna()

if len(recent_hitters) > 0 and len(mid_hitters) > 0 and len(early_hitters) > 0:
    f_stat, p_value_anova = stats.f_oneway(recent_hitters, mid_hitters, early_hitters)
    print(f"\nANOVA Test (Hitters WAR across eras):")
    print(f"  F-statistic: {f_stat:.3f}")
    print(f"  P-value: {p_value_anova:.4f}")
    if p_value_anova < 0.05:
        print(f"  Result: Significant difference between eras (p < 0.05)")
    else:
        print(f"  Result: No significant difference between eras (p >= 0.05)")

if not pitchers_df.empty:
    pitchers_df['Era'] = pitchers_df['Contract_Start'].apply(
        lambda x: 'Recent (2023+)' if x >= 2023 else 'Mid (2015-2018)' if x >= 2015 else 'Early (2007-2010)'
    )
    
    print("\n\nPitchers - WAR by Era:")
    era_groups_pitchers = pitchers_df.groupby('Era')['WAR']
    print(era_groups_pitchers.describe())
    
    recent_pitchers = pitchers_df[pitchers_df['Era'] == 'Recent (2023+)']['WAR'].dropna()
    mid_pitchers = pitchers_df[pitchers_df['Era'] == 'Mid (2015-2018)']['WAR'].dropna()
    early_pitchers = pitchers_df[pitchers_df['Era'] == 'Early (2007-2010)']['WAR'].dropna()
    
    if len(recent_pitchers) > 0 and len(mid_pitchers) > 0 and len(early_pitchers) > 0:
        f_stat_p, p_value_anova_p = stats.f_oneway(recent_pitchers, mid_pitchers, early_pitchers)
        print(f"\nANOVA Test (Pitchers WAR across eras):")
        print(f"  F-statistic: {f_stat_p:.3f}")
        print(f"  P-value: {p_value_anova_p:.4f}")
        if p_value_anova_p < 0.05:
            print(f"  Result: Significant difference between eras (p < 0.05)")
        else:
            print(f"  Result: No significant difference between eras (p >= 0.05)")
```

```python
print("\n" + "-" * 60)
print("Statistical Test 3: High AAV vs Low AAV Performance")
print("-" * 60)

hitter_median_aav = hitters_df['AAV (M)'].median()
hitters_df['AAV_Group'] = hitters_df['AAV (M)'].apply(
    lambda x: 'High AAV' if x >= hitter_median_aav else 'Low AAV'
)

high_aav_hitters = hitters_df[hitters_df['AAV_Group'] == 'High AAV']['WAR']
low_aav_hitters = hitters_df[hitters_df['AAV_Group'] == 'Low AAV']['WAR']

print(f"\nHitters (Median AAV: ${hitter_median_aav:.2f}M):")
print(f"  High AAV group - Mean WAR: {high_aav_hitters.mean():.2f}, Std: {high_aav_hitters.std():.2f}")
print(f"  Low AAV group - Mean WAR: {low_aav_hitters.mean():.2f}, Std: {low_aav_hitters.std():.2f}")

t_stat_h, p_value_t_h = stats.ttest_ind(high_aav_hitters, low_aav_hitters)
print(f"\n  T-test results:")
print(f"    T-statistic: {t_stat_h:.3f}")
print(f"    P-value: {p_value_t_h:.4f}")
if p_value_t_h < 0.05:
    print(f"    Result: Significant difference (p < 0.05)")
else:
    print(f"    Result: No significant difference (p >= 0.05)")

if not pitchers_df.empty:
    pitcher_median_aav = pitchers_df['AAV (M)'].median()
    pitchers_df['AAV_Group'] = pitchers_df['AAV (M)'].apply(
        lambda x: 'High AAV' if x >= pitcher_median_aav else 'Low AAV'
    )
    
    high_aav_pitchers = pitchers_df[pitchers_df['AAV_Group'] == 'High AAV']['WAR']
    low_aav_pitchers = pitchers_df[pitchers_df['AAV_Group'] == 'Low AAV']['WAR']
    
    print(f"\n\nPitchers (Median AAV: ${pitcher_median_aav:.2f}M):")
    print(f"  High AAV group - Mean WAR: {high_aav_pitchers.mean():.2f}, Std: {high_aav_pitchers.std():.2f}")
    print(f"  Low AAV group - Mean WAR: {low_aav_pitchers.mean():.2f}, Std: {low_aav_pitchers.std():.2f}")
    
    t_stat_p, p_value_t_p = stats.ttest_ind(high_aav_pitchers, low_aav_pitchers)
    print(f"\n  T-test results:")
    print(f"    T-statistic: {t_stat_p:.3f}")
    print(f"    P-value: {p_value_t_p:.4f}")
    if p_value_t_p < 0.05:
        print(f"    Result: Significant difference (p < 0.05)")
    else:
        print(f"    Result: No significant difference (p >= 0.05)")
```
